```python
brvehins_ls = [
    "brvehins1a.rda", 
    "brvehins1b.rda", 
    "brvehins1c.rda", 
    "brvehins1d.rda", 
    "brvehins1e.rda"
]
df_list=[]

for dataset in brvehins_ls:
    data_dict = pyreadr.read_r(str(data_dir / dataset))
    # print(data_dict)
    df = data_dict[dataset.split(".")[0]]
    df_list.append(df)

df=pd.concat(df_list, ignore_index=True)
```

```python
input_df["VehYear"] = input_df["VehYear"].astype(int)
```

```python
for col in ["Gender", "DrivAge"]:
    input_df[col] = input_df[col].astype(str)
    input_df[col] = input_df[col].fillna(value = "No information")
```

```python
for col in ["VehModel", "VehGroup"]:
    #input_df[col] = input_df[col].astype(str)
    input_df[col] = input_df[col].fillna(value = "Outros")
```

```python
input_df.isna().sum()
```

```python
conditions = (
    (input_df["VehYear"] > 0)
    & (input_df["ExposTotal"] > 0)
    & (input_df["PremTotal"] > 0)
    & (input_df["SumInsAvg"] > 0)
)
```

```python
input_df.shape
```

```python
df = input_df[conditions]
df.shape
```

create frequency, severity and pure premium columns

```python
claim_types = ["Rob", "PartColl", "TotColl", "Fire", "Other"]

for claim_type in claim_types:
    claim_nb_col = f"ClaimNb{claim_type}"
    claim_amount_col = f"ClaimAmount{claim_type}"

    # Frequemncy = ClaimNb / ExposTotal
    df[f"Freq{claim_type}"] = df[claim_nb_col] / df["ExposTotal"]

    # Severity = ClaimAmount / ClaimNb (0 if no claims)
    df[f"Sev{claim_type}"] = np.where(
        df[claim_nb_col] > 0,
        df[claim_amount_col] / df[claim_nb_col],
        0
    )

    # Pure Premium = ClaimAmount / ExpostTotal
    df[f"PP{claim_type}"] = df[claim_amount_col] / df["ExposTotal"]
```
