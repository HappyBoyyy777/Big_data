```python
gdp_per_capita_2020 = gdp_per_capita[gdp_per_capita["Year"]==2020]
gdp_per_capita_2020=gdp_per_capita_2020.rename(
    columns={"GDP per capita, PPP (constant 2017 international $)": "GDP per capita (USD)"}
)
gdp_per_capita_2020
```

```python
#pivot table
#TOT - Total

oecd_bli = oecd_bli[oecd_bli["INEQUALITY"]=="TOT"]
oecd_bli = oecd_bli.pivot(index="Country",columns="Indicator",values="Value")
oecd_bli
```

```python
#Chcemy, żeby kraj wrócił do kolumn

oecd_bli = oecd_bli.reset_index()
oecd_bli
```

```python
# do metody merge, w obu tabelach muszą być klucze tego samego typu (w naszym przypadku dtype to object, więc się uda)
# lewa strona - oecd, prawa gdp_per_capita => dołączamy klucze do tabeli przy użyciu klucza
# merge podobna do metody join
# left/right/(full)outer/inner join
# np. inner join pozbywa się NULLi

full_country_stats = oecd_bli.merge(gdp_per_capita_2020, left_on="Country",right_on="Entity",how="inner")

full_country_stats
```

```python
# ograniczenie do 3 kolumn

full_country_stats = full_country_stats[["Country","Life satisfaction","GDP per capita (USD)"]]
full_country_stats = full_country_stats.sort_values(by="GDP per capita (USD)")
full_country_stats
```

```python
# index = false po to żeby pozbyć się zbędnych indeksow

full_country_stats.to_csv("D:\python_projekty\pum\data\lifesat\lifesat_full.csv",index=False)
```

```python
min_gdp = 23_500
max_gdp = 62_500

country_stats = full_country_stats[(full_country_stats["GDP per capita (USD)"] >= min_gdp) & (full_country_stats["GDP per capita (USD)"] <= max_gdp)]
country_stats.to_csv("D:\python_projekty\pum\data\lifesat\lifesat.csv",index=False)
```

```python
#Wczytanie danych
lifesat = pd.read_csv("D:\python_projekty\pum\data\lifesat\lifesat.csv")
#Przygotowanie danych - inplace=True nadpisuje zmienną lifesat - trzeba wyswietlic dataframe, zeby sprawdzic czy wszystko jest w porzadku
lifesat.rename(columns={"GDP per capita (USD)":"PKB na mieszkanca (USD)","Life satisfaction":"Satysfakcja z zycia"},inplace=True)
lifesat
```

```python
#wizualizujacja danych
lifesat.plot(kind='scatter', grid=True,x="PKB na mieszkanca (USD)",y="Satysfakcja z zycia")
plt.show()
```

Dane są zaszumione, ale wygląda na to, że poziom satysfakcji z życia wzrasta w sposób mniej więcej liniowy wraz ze wzrostem PKB.

Faza trenowania modelu:

```python
#Przygotowanie zmiennych w modelu
X = lifesat[["PKB na mieszkanca (USD)"]]
y = lifesat[["Satysfakcja z zycia"]]

#Wybranie modelu liniowego
model = LinearRegression()

#Trenowanie modelu
model.fit(X,y)

#Obliczenie współczynników modelu
theta_0=model.intercept_[0]
theta_1=model.coef_[0][0]
print(f"theta_0: {theta_0}")
print(f"theta_1: {theta_1}")
```

```python
#wizualizacja danych - C.D
lifesat.plot(kind='scatter', grid=True,x="PKB na mieszkanca (USD)",y="Satysfakcja z zycia")
plt.plot(X,theta_0+theta_1*X)
plt.text(45000,6.5,fr"$\theta_0 = {theta_0.round(2)}$", color="blue")
plt.text(45000,6.3,fr"$\theta_1 = {theta_1 * 1e5:.2f}\cdot 10^{{-5}}$", color="blue")
plt.title("Zależność między PKB, a satysfakcją z życia")
plt.show()
```

Wykorzystujemy model, żeby sprawdzić jak szczęśliwi są Cypryjczycy

```python
cyprus_gdp_per_capita_2020 = (gdp_per_capita_2020[gdp_per_capita_2020["Entity"]=="Cyprus"][["GDP per capita (USD)"]])
cyprus_gdp_per_capita_2020.rename(columns={"GDP per capita (USD)":"PKB na mieszkanca (USD)"},inplace=True)
cyprus_gdp_per_capita_2020
```

```python
cyprus_predicted_life_satisfaction = model.predict(cyprus_gdp_per_capita_2020)

cyprus_predicted_life_satisfaction
```
